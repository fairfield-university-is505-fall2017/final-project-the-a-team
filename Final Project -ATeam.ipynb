** Which types of household have the highest incomes?**

```python
# create an empty list 
t_list = []
index = []
df = result['marital.status'].drop_duplicates()

# Run the for loop to add the values in the t_list
for i, status in df.iteritems():
    count = result.groupby(by=result['marital.status'] == status)['income'].value_counts()
    data = count.tolist()[2:4]
    t_list.append(data)
    index.append(status)
    
# create the new dateframe from t_list   
df = pd.DataFrame(t_list).T
df.columns = index 

# create pie charts to check the connection between marital status and income
plt.style.use('ggplot')
labels = '<=50K', '>50K'
fig, axes = plt.subplots(1,7, figsize=(20,4))

for ax, col in zip(axes, df.columns):
    ax.pie(df[col], labels=labels, autopct='%1.1f%%', startangle=90, shadow=True)
    ax.set(ylabel='', title=col, aspect='equal')
plt.show()
```

For the marital status category, married-civ-spouse (married, with civilian spouse) earns more than other statuses. 44.7% of married civ-spouse earn more than 50k per year.

**Map displaying the total amount of immigrants comming to the US from other countries:**

```python
# Initialize an empty dictionary: counts_dict
counts_dict = {}
# Iterate over the column in DataFrame
for entry in result['native.country']:
    if entry in counts_dict.keys():
        counts_dict[entry] += 1
    else:
        counts_dict[entry] = 1
del counts_dict['?']
del counts_dict['United-States']
del counts_dict['Outlying-US(Guam-USVI-etc)']

# Print the populated dictionary
print(counts_dict)
```

```python
# draw World map
plt.figure(figsize=(12,6))
m = Basemap()
m.drawcoastlines()
m.drawcountries()
m.fillcontinents(color='gray',lake_color='aqua')

# draw parallels and meridians.
m.drawparallels(np.arange(-90.,120.,30.))
m.drawmeridians(np.arange(0.,360.,60.))
m.drawmapboundary(fill_color='aqua')
geolocator = Nominatim()

# Get the location of each city and plot it
scale = 1
for (country,count) in counts_dict.items():
    location = geolocator.geocode(country)
    if location is not None and location.longitude is not None:
        x, y = m(location.longitude, location.latitude)
    m.plot(x,y,marker='o',color='Red',markersize=int(math.sqrt(count))*scale)
    
plt.show()
```

Mexico is the country with largest number of immigrants to the US. In our dataset, people come from The Caribbean, Europe, and Asia. Of the people in the dataset, none came from Australasia or Africa. 

**After looking at the world map, we wondered which states the immigrants reside in. So, 
we decided to find another dataset about the number of immigrants in each states, and use 
the heatmap to figure out our answer. This is the second, crossover part of our analysis!**

```python
# Loading the dataset into the dataframe 
# The dataset was taken from https://www.dhs.gov/immigration-statistics/yearbook/2015/table4
source = pd.read_excel('fy2015_table4.xls',header=0)
# Zipping two columns together and create a dictionary 
states_2015 = zip(source['State or territory of residence'],source[2015])
states_dict = dict(states_2015)
del states_dict['- Represents zero.']
del states_dict['1 Includes American Samoa, Northern Mariana Islands, U.S. Virgin Islands, and U.S. Armed Forces posts.']
del states_dict['Source: U.S. Department of Homeland Security.']
# states_dict
```

```python
# Lambert Conformal map of lower 48 states.
plt.figure(figsize=(14,7))
m = Basemap(llcrnrlon=-119,llcrnrlat=22,urcrnrlon=-64,urcrnrlat=49,
        projection='lcc',lat_1=32,lat_2=45,lon_0=-95)
shp_info = m.readshapefile('st99_d00','states',drawbounds=True)
fig = plt.gcf()
states_dict
cmap = plt.cm.plasma 
vmin = 0; vmax = 250000 # set range.
norm = mpl.colors.Normalize(vmin=vmin, vmax=vmax)
mapper = ScalarMappable(norm=norm, cmap=cmap)
# choose a color for each state based on population density.
colors={}
statenames=[]
for shapedict in m.states_info:
    statename = shapedict['NAME']
    pop = states_dict[statename]
    colors[statename] = mapper.to_rgba(pop)
    statenames.append(statename)
    
# cycle through state names, color each one.
ax = plt.gca() # get current axes instance
for nshape,seg in enumerate(m.states):
    color = rgb2hex(colors[statenames[nshape]]) 
    poly = Polygon(seg,facecolor=color,edgecolor=color)
    ax.add_patch(poly)    
plt.title('Total number of Immigrants Residing in Each State, Continental US, 2015')
cax = fig.add_axes([0.115, 0.06, 0.8, 0.04]) # posititon
cb = ColorbarBase(cax,cmap=cmap,norm=norm, orientation='horizontal')

plt.show()
```

This revealed an interesting trend: California has the most immigrants of any state, at least of 2015. Following California, the states which other high numbers of immigrants is New York, Texas, Florida, and Illinois, in descending order.
